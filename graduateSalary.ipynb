```python
#For intake
def field2(n,cat,calc,year):
    fieldlUnique = np.unique(n['Field'])
    years = np.unique(n[year])
    coursesInfo = {
    'Business':[],
    'Arts':[],
    'Business':[],
    'Medicine':[],
    'Education':[],
    'Engineering':[],
    'Arts':[],
    'Medicine':[],
    'Arts':[],
    'Computing':[],
    'Law':[],
    'Business':[],
    'Medicine':[],
    'Science':[],
    }
    for j in years:
        for i in range(len(fieldlUnique)):
            if fieldlUnique[i] == 'nil':
                continue
            data = n[(n['Field'] == fieldlUnique[i]) & (n[year] == j)][cat]
            if calc == 'mean':

                coursesInfo[fieldlUnique[i]].append(np.mean(data))
            elif calc == 'sum':
                coursesInfo[fieldlUnique[i]].append(np.sum(data))
            else:
                 coursesInfo[fieldlUnique[i]].append((data))
    return coursesInfo
```

```python
#filter  by insituition
def fieldByInsitute(n,cat):
    fieldlUnique = np.unique(n['Field'])
    insitute = np.unique(graduateSalarySG['Institution'])
    coursesInfo = {
    'Business':[],
    'Arts':[],
    'Business':[],
    'Medicine':[],
    'Education':[],
    'Engineering':[],
    'Arts':[],
    'Medicine':[],
    'Arts':[],
    'Computing':[],
    'Law':[],
    'Business':[],
    'Medicine':[],
    'Science':[],
    "Human Resource":[],
    }
    for j in insitute:
        for i in range(len(fieldlUnique)):
            if fieldlUnique[i] == 'nil':
                continue
            
            data = n[(n['Field'] == fieldlUnique[i]) & (n['Institution'] == j) & (n['Year_of_Survey'] ==2020)][cat]
            coursesInfo[fieldlUnique[i]].append(np.mean(data))
    return coursesInfo
```
